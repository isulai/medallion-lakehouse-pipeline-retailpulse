
###THis is the Bronze layer for our Project

Gold Table 1 — Revenue by City

```python
spark.sql("USE CATALOG workspace")
spark.sql("USE SCHEMA retailpulse_project")
```

```python
# Read Silver table

silver_df = spark.table("workspace.retailpulse_project.silver_sales_enriched")
```

```python
# Create Aggregation:

from pyspark.sql.functions import sum

gold_city_revenue_df = (
    silver_df.groupBy("city")
    .agg(sum("revenue").alias("total_revenue")
         )
)
```

```python
display(gold_city_revenue_df)
```

```python
# Save the 1st Gold Table

gold_city_revenue_df.write.format("delta") \
    .mode("overwrite") \
        .saveAsTable(
            "workspace.retailpulse_project.gold_city_revenue"
        )
```

```python
%sql

select * from workspace.retailpulse_project.gold_city_revenue;
```


Gold 2 - Top Customers

```python
from pyspark.sql.functions import col

gold_top_customers_df = (
    silver_df.groupBy("name") \
        .agg(
            sum("revenue").alias("total_revenue")
        )
        .orderBy(col("total_revenue").desc()
        )
)
```

```python
display(gold_top_customers_df)
```

```python
# Save as a table

gold_top_customers_df.write.format("delta") \
    .mode("overwrite") \
        .saveAsTable(
            "workspace.retailpulse_project.gold_top_customers"
        )
```

```python
%sql

SELECT * From workspace.retailpulse_project.gold_top_customers;
```


Gold 3 - Product Revenue

```python
gold_product_revenue_df = (
    silver_df.groupBy("product_name", "category") \
        .agg(sum("revenue").alias("total_revenue")) \
            .orderBy(col("total_revenue").desc()
                     )
)
```

```python
gold_product_revenue_df.show()
```

```python
#Saving the Table for Products.

gold_product_revenue_df.write.format("delta") \
    .mode("overwrite") \
        .saveAsTable(
            "workspace.retailpulse_project.gold_product_revenue"
        )
```

```python
%sql

SELECT * FROM workspace.retailpulse_project.gold_product_revenue;
```
