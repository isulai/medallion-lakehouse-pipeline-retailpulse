
## THis is the Bronze layer for our Project

Step 1 — Read Customers CSV

```python
spark.sql("USE CATALOG workspace")
spark.sql("USE SCHEMA retailpulse_project")
```

```python
from pyspark.sql.functions import current_timestamp

customer_df = customer_df = (
    spark.read
    .option("header", "true")
    .csv("/Volumes/workspace/retailpulse_project/retailpulse_files/customers/customers.csv")
)

display(customer_df)
```

```python
#Adding the timestamp column adding in the customer_df

customer_df = customer_df.withColumn(
    "ingestion_timestamp", current_timestamp()
)

display(customer_df)
```

Step 2 — Save to Bronze Table

```python
%sql
DROP TABLE workspace.retailpulse_project.bronze_customers
```

```python
customer_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "workspace.retailpulse_project.bronze_customers"
    )
```

```python
%sql
DESCRIBE TABLE workspace.retailpulse_project.bronze_customers
```


Now for Products & Orders we do the same

```python
from pyspark.sql.functions import current_timestamp

product_df = (
    spark.read
    .option("header","true")
    .csv("/Volumes/workspace/retailpulse_project/retailpulse_files/products/products.csv")
)

product_df = product_df.withColumn(
    "ingestion_timestamp",
    current_timestamp()
)

product_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "workspace.retailpulse_project.bronze_products"
    )
```

```python
%sql

DESCRIBE TABLE workspace.retailpulse_project.bronze_products;
```

```python
from pyspark.sql.functions import current_timestamp

order_df = (
    spark.read
    .option("header","true")
    .csv("/Volumes/workspace/retailpulse_project/retailpulse_files/orders/orders.csv")
)

order_df = order_df.withColumn(
    "ingestion_timestamp",
    current_timestamp()
)

order_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "workspace.retailpulse_project.bronze_orders"
    )
```

```python
%sql

DESCRIBE TABLE workspace.retailpulse_project.bronze_orders;
```
